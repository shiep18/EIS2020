# 用opencv在某个区域查找垃圾箱并标上红框

```python
import cv2
import numpy as np
import pyautogui as pg

pic=pg.screenshot(region=(1150,450, 200, 200))
img = cv2.cvtColor(np.array(pic),cv2.COLOR_RGB2BGR)
template = cv2.imread('recyclebin.PNG')
(h, w )= template.shape[:2]

res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.9

loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

cv2.imshow("img",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# 找到所有的x并加红框

```python
import cv2
import numpy as np
import pyautogui as pg

pic=pg.screenshot()
img = cv2.cvtColor(np.array(pic),cv2.COLOR_RGB2BGR)
template = cv2.imread('xb.PNG')
(h, w )= template.shape[:2]

res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.9

loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
resimg=cv2.resize(img,(300,200))
cv2.imshow("img",resimg)
cv2.waitKey()
cv2.destroyAllWindows()
```

# 找到所有的x,点击自动关闭窗口

```python
import cv2
import numpy as np
import pyautogui as pg
import time

pic=pg.screenshot()
img = cv2.cvtColor(np.array(pic),cv2.COLOR_RGB2BGR)
template = cv2.imread('xb.PNG')
(h, w )= template.shape[:2]

res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.9

loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    print("moveto",pt[0],pt[1])
    time.sleep(1)
    pg.moveTo(pt[0]+w/2,pt[1]+h/2)
    time.sleep(1)
    pg.click()
```
